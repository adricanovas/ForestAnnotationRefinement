```python
import sys
from pathlib import Path
import os
import laspy
import numpy as np
import pandas as pd
import time 
import torch
```

```python
def process_pth(pth):
    x, y, z = pth['coord'][:,0], pth['coord'][:,1], pth['coord'][:,2]

    label = np.full(len(x), -1, dtype=np.int32)

    confidences = pth['pred_scores']

    for i in range(pth['pred_masks'].shape[0]):
        tree = pth['pred_masks'][i]
        mask = tree == 1
        label[mask] = i

    trees_df = pd.DataFrame({
        'x': x,
        'y': y,
        'z': z,
        'label': label
    })

    return trees_df, confidences
```

# Conference Paper Results

```python
from GlobalEvaluator2 import GlobalEvaluator, compute_tile_metrics
root = Path("../..").resolve()
sys.path.append(str(root))
```

```python
models = ["OACNN_0", "OACNN_25", "OACNN_50", "OACNN_100", "OACNN_ML"]
base_path_preds = "../../data/Santomera/result"
base_path_gt = "../../data/Santomera/tiles/trees/labeled"

for model in models:
    evaluator = GlobalEvaluator()
    print(f"Processing model: {model}")
    model_path = os.path.join(base_path_preds, model)
    for file in os.listdir(model_path):
        if file.endswith("_pred.pth"):
            pred_result_path = os.path.join(model_path, file)
            results = torch.load(pred_result_path, map_location=torch.device('cpu'))
            trees_df, confidences = process_pth(results)
            gt_file = file.replace("_pred.pth", ".las")
            gt_path = os.path.join(base_path_gt, gt_file)
            tile_result = compute_tile_metrics(trees_df, gt_path, confidences)
            evaluator.accumulate_tile(tile_result)
    print(evaluator.report_micro_global())
    print("---" * 20)
```
